# Data Cleaning

### Drop columns that do not provide substantial information
- Timezone, State, Country are not useful for analyzing California data
- We have the accident start time, so Twilight/Sunrise_Sunset data are not useful
- Airport_Code shows the source of weather info, not particularly useful
- End_Lat, End_Lng contain a significant amount of NaN values, not useful

```python
cadf[cadf['Timezone'] == 'US/Mountain'].count()
#there are 44 accidents logged using MST. Convert these to PST or drop?
```

```python
drop_cols = ['End_Lat', 'End_Lng', 'Number', 'TMC', 'Zipcode', 'Country', 'Timezone',
            'Airport_Code', 'Side', 'State', 'Civil_Twilight', 'Wind_Chill(F)',
            'Nautical_Twilight', 'Astronomical_Twilight', 'ID', 'Source',
            'Wind_Direction', 'Weather_Timestamp']
ccadf = cadf.drop(columns=drop_cols)
#cleaned cadf = ccadf
```

```python
#Percentage of NaN values in each column
ccadf.isna().sum()/len(ccadf)
```

### Fill Na Values in `ccadf`:
- 2.4% of Temperature are NaN --> fill with mean temp
- 59.4% of Precipitation are NaN --> fill with 0, meaning no rain
- 2.6% of Humidity are NaN --> fill with mean
- 1.8% of Pressure are NaN --> fill with mean
- 2% of visibility are NaN --> What to do here...? 16.5k rows... Fill with mean for now
- 16% of Wind Speed are NaN --> fill with mean
- 2% of Weather Condition are NaN --> Drop these rows? 16.3k rows total...
- only 8 rows of Sunrise_Sunset = NaN --> Drop

```python
ccadf['Temperature(F)'] = ccadf['Temperature(F)'].fillna(ccadf['Temperature(F)'].mean())
ccadf['Precipitation(in)'] = ccadf['Precipitation(in)'].fillna(0)
ccadf['Humidity(%)'] = ccadf['Humidity(%)'].fillna(ccadf['Humidity(%)'].mean())
ccadf['Pressure(in)'] = ccadf['Pressure(in)'].fillna(ccadf['Pressure(in)'].mean())
ccadf['Visibility(mi)'] = ccadf['Visibility(mi)'].fillna(ccadf['Visibility(mi)'].mean())
ccadf['Wind_Speed(mph)'] = ccadf['Wind_Speed(mph)'].fillna(ccadf['Wind_Speed(mph)'].mean())
ccadf = ccadf[ccadf['Weather_Condition'].notna()]
ccadf = ccadf[ccadf['Sunrise_Sunset'].notna()]
```

```python
ccadf.info()
#Went from 816k rows to 800k rows - this should still be sufficient data for valuable prediction
```

```python
#Double check that there are no more NaN values in the DF
ccadf.isna().sum()/len(ccadf)
```

### Change `Start_Time` and `End_Time` to `datetime` format
- Parse Time and Date from the column into their own respective columns
- Delete the original column that combined the date and time info

```python
ccadf['Start_TS'] = pd.to_datetime(ccadf['Start_Time'])
ccadf['End_TS'] = pd.to_datetime(ccadf['End_Time'])

ccadf['Start_Time'] = ccadf['Start_TS'].dt.time
ccadf['End_Time'] = ccadf['End_TS'].dt.time
ccadf['Date'] = ccadf['Start_TS'].dt.date

ccadf = ccadf.drop(columns=['Start_TS', 'End_TS'])
```

```python
ccadf['Start_Time'].head()
```

### Convert boolean column values from True and False to 1 and 0, respectively

```python
bool_cols = ['Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway',
       'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal',
       'Turning_Loop']
ccadf[bool_cols] = (ccadf[bool_cols] == True).astype(int)
```

#### Let's look at how different Weather Conditions are correlated with Severity
- There are 72 total weather conditions (Clear, fair, mostly cloudy, partly cloudy, overcast, etc.)
    - some are combinations of one another (drizzle/windy, windy/snowy, etc), some are varying in weather severity (partly, mostly, light, heavy, etc)


- Instinctively, worse weather condition should lead to higher accident severity, but let's find out

```python
#How many accidetns under each Weather Condition?
ccadf['Weather_Condition'].value_counts().plot(figsize=(10,4), 
                                               title='Number of Accidents by Weather Condition', rot=90,
                                              ylim=(0,100))

#Some of the more extreme weather conditions have little data... Where do I set the cutoff?
```

```python
wcdf = pd.DataFrame()
for condition in ccadf['Weather_Condition'].unique():
#     print('WC', condition,
#           cadf[cadf['Weather_Condition'] == condition]['ID'].count(), 'Severity',
#          cadf[cadf['Weather_Condition'] == condition]['Severity'].mean())
    wcdf = wcdf.append({"Weather_Condition": f'{condition}',
                 "WC_Count": float(f"{cadf[cadf['Weather_Condition'] == condition]['ID'].count()}"),
                  "Mean_Severity": float(f"{cadf[cadf['Weather_Condition'] == condition]['Severity'].mean()}")},
               ignore_index=True)
```

```python
fig, ax = plt.subplots(figsize=(13,4))
# wcdf.plot(x='Weather_Condition', y='Mean_Severity', ax=ax, rot=90, x_compat=True,
#          title='Weather Condition vs. Mean Accident Severity')
# plt.xticks(wcdf.Weather_Condition)
ax.plot(wcdf['Weather_Condition'], wcdf['Mean_Severity']);
plt.xticks(rotation='vertical');
```

### There seems to be a lot of Weather Conditions that have very little data. 
Let's filter them out by setting the minimum count to 50

```python
test = wcdf.sort_values(by=['WC_Count'], ascending=False)
test.tail()
```

```python
#Create new DF that contains weather conditions for which the count > 100
wcdf1 = wcdf[wcdf['WC_Count'] > 50]
wcdf1 = wcdf1.sort_values(by=['Mean_Severity'], ascending=False)

#Total of 35 weather conditions were filtered out, but which ones?
wcdf[wcdf['WC_Count'] <= 50]['Weather_Condition'].unique()
```

```python
fig, ax = plt.subplots(figsize=(12,4))
# wcdf.plot(x='Weather_Condition', y='Mean_Severity', ax=ax, rot=90, x_compat=True,
#          title='Weather Condition vs. Mean Accident Severity')
# plt.xticks(wcdf.Weather_Condition)
ax.plot(wcdf1['Weather_Condition'], wcdf1['Mean_Severity']);
plt.xticks(rotation='vertical');
plt.title("Mean Accident Severity vs. Weather Condition")
ax.set_xlabel("Weather Condition")
ax.set_ylabel("Mean Severity on Scale of 1-5")
```

### There does not seem to be a very clear relationship between Accident Severity and Weather Condition:
- Clear seems to be worse than T-Storm... Maybe people tend to drive more safely during a T-Storm?
- Top 10 Weather Conditions in terms of Accident Severity:
    1. Rain Showers
    2. Shallow Fog
    3. Thunderstorms and Rain
    4. Drizzle
    5. Light Drizzle
    6. Scattered Clouds
    7. Overcast
    8. Clear
    9. Light Thunderstorms and Rain
    10. Haze
